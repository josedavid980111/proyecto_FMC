```python
import re, string
archivo = input("Nombre del archivo: ")
f=open(archivo+".txt").read()
#f=open("Ejemplo.txt").read()

#float y double
Aux1=re.sub(r"[.]","", f)
Aux2=re.sub("\s",",", Aux1)
Aux=re.sub('"', '', Aux2)
```

```python
ExpReg11 = "int"+r",+"+"\w+"
Exp11 = re.findall(ExpReg11,Aux,flags=re.MULTILINE)

ExpReg12 = "int"+r",+"+"\w+"+r",*"+"=*"
Exp12 = re.findall(ExpReg12,Aux,flags=re.MULTILINE)

bandera=True
banderaF=False
inicio=0
while banderaF==False:
    if len(Exp11)==len(Exp12):
        x=0
        while x < len(Exp11) and bandera==True:
            
            Aux1=Exp11[x]
            Aux2=Exp12[x]
            if(Aux1!=Aux2):
                bandera=False
            if x+1 == len(Exp11) and bandera==True :
                banderaF=True
            x+=1
        bandera=True
            
            
    if banderaF==False and inicio==0:
        ExpReg11=ExpReg11+r",*"+"=*"
        ExpReg12=ExpReg12+"\w*"+r",*"+"=*"
        inicio+=1
    elif banderaF==False and inicio==1:
        ExpReg11=ExpReg11+"\w*"+r",*"+"=*"
        ExpReg12=ExpReg12+"\w*"+r",*"+"=*"
    
    Exp11 = re.findall(ExpReg11,Aux,flags=re.MULTILINE)
    Exp12 = re.findall(ExpReg12,Aux,flags=re.MULTILINE)
    
#print(Exp12)

ResInt=[]
for x in range(len(Exp12)):
    Au=re.sub(r"int,*","", Exp12[x])
    A=re.sub(r",*"+r"="+r",*"+"\w*"+r",*"," ", Au)
    ExpReg14= "\w+"
    Exp14 = re.findall(ExpReg14,A)
    for y in range(len(Exp14)):
        ResInt.append(Exp14[y])
print(ResInt)

#for x in range(len(Exp2)):
    


#+"\w*"+r",*"+"=*"
```
